```
import json
movies_df = (spark
             .read
             .json("dbfs:/FileStore/movielens/test-data/movies.json")
             .drop('genres')
             .where('movieId IN (9999, 6548)')
             .cache())
```

```
movies_list = as_list(movies_df)
```

```
print(json.dumps(movies_list, indent=2))
```

```
from operator import itemgetter
movies_list_sorted = as_list(movies_df, order_by=itemgetter("movieId"))
```

```
print(json.dumps(movies_list_sorted, indent=2))
```

```
expect = [
  { "movieId": 6548, "title": "Bad Boys II (2003)", },
  { "movieId": 9999, "title": "Unknow within", },
]

import unittest
test_case = unittest.TestCase()
test_case.assertEqual(expect, movies_list_sorted)
```
